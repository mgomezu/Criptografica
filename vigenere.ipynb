```python
import numpy as np

def encriptar():
  A = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
  mensaje = input("Escriba el mensaje: ")
  k = input("Escriba la llave: ")
  t = input("Escriba el tamaño del bloque: ")
  mensaje = mensaje.replace(' ','')
  mensaje = mensaje.lower()
  plaintext = []
  key = []
  codificado = ""
  for letra in mensaje:
    plaintext.append(A.index(letra))

  for letra in k:
    key.append(A.index(letra))

  i=0
  for caracteres in plaintext:

    cod = (int(caracteres) + int(key[i%len(key)])) % 26
    letter = A[cod]
    codificado = codificado + letter
    i += 1
    if i % int(t) == 0:
      codificado = codificado + " "

  print(codificado)

def desencriptar():
  A = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
  mensaje = input("Escriba el mensaje: ")
  k = input("Escriba la llave: ")
  t = input("Escriba el tamaño del bloque: ")
  mensaje = mensaje.replace(' ','')
  mensaje = mensaje.lower()
  plaintext = []
  key = []
  codificado = ""
  for letra in mensaje:
    plaintext.append(A.index(letra))

  for letra in k:
    key.append(A.index(letra))

  i=0
  for caracteres in plaintext:

    cod = (int(caracteres) - int(key[i%len(key)])) % 26
    letter = A[cod]
    codificado = codificado + letter
    i += 1
    if i % int(t) == 0:
        codificado = codificado + " "
  print(codificado)

try:
  opcion = 1
  while opcion != 0:
    opcion = input("Escriba 1 para codificar, 2 para decodificar o 0 para terminar: ")
    if int(opcion) == 0:
      break
    elif int(opcion) == 1:
      encriptar()
    elif int(opcion) == 2:
      desencriptar()
    else:
      print("Oops!  Numero no valido.")


except ValueError:
  print("Oops!  Numero no valido.")
```
