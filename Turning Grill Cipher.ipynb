```python
from math import sqrt

def cifrar(mensaje, cuadricula):

    cifrado = []
    pos = 0

    for i in range(len(cuadricula)):
      fila = []
      for j in range(len(cuadricula[i])):
          fila.append(0)
      cifrado.append(fila)

    for n in range(4):
      for i in range(len(cuadricula)):
        for j in range(len(cuadricula[i])):
            if cuadricula[i][j] == 1:
              if cifrado[i][j] == 0:
                cifrado[i][j] = mensaje[pos]
                pos+=1

      cuadriculaAux = []
      for i in range(len(cuadricula)):
        fila = []
        for j in range(len(cuadricula[i])):
          fila.append(cuadricula[j][len(cuadricula)-i-1])
        cuadriculaAux.append(fila)

      cuadricula = cuadriculaAux


    return cifrado

def descifrar(cifrado, cuadricula):
    descifrado = ""
    Aux = []

    for i in range(len(cuadricula)):
      fila = []
      for j in range(len(cuadricula[i])):
          fila.append(0)
      Aux.append(fila)
    for n in range(4):
      for i in range(len(cuadricula)):
        for j in range(len(cuadricula[i])):
            if cuadricula[i][j] == 1:
              if Aux[i][j] == 0:
                descifrado += cifrado[i][j]
                Aux[i][j] = 1

      cuadriculaAux = []
      for i in range(len(cuadricula)):
        fila = []
        for j in range(len(cuadricula[i])):
          fila.append(cuadricula[j][len(cuadricula)-i-1])
        cuadriculaAux.append(fila)

      cuadricula = cuadriculaAux
    return descifrado

def main():
    try:
        opcion = 1
        while opcion != 0:
            opcion = input("Escriba 1 para cifrar, 2 para descifrar o 0 para terminar: ")
            if int(opcion) == 0:
                break
            elif int(opcion) == 1:
                cuadricula = []
                mensaje = input("Escriba el mensaje: ")
                mensajeN = mensaje.replace(" ", "")
                k= input("Escriba la llave: ")
                matrix = k.split()
                matrixNumeros = [int(cadena) for cadena in matrix]
                raiz = int(sqrt(len(matrixNumeros)))
                for i in range(raiz):
                    fila = matrixNumeros[i * raiz : (i + 1) * raiz]
                    cuadricula.append(fila)
                cifrado = cifrar(mensajeN, cuadricula)
                print("Mensaje cifrado: \n")
                for fila in cifrado:
                  print(fila)

            elif int(opcion) == 2:
                cuadricula = []
                cifrado = input("Escriba el mensaje cifrado: ")
                matrix = cifrado.split()
                raiz = int(sqrt(len(matrix)))
                cuadriculaCifrada = []
                for i in range(raiz):
                    fila = matrix[i * raiz : (i + 1) * raiz]
                    cuadriculaCifrada.append(fila)
                k= input("Escriba la llave: ")
                matrixN = k.split()
                matrixNumeros = [int(cadena) for cadena in matrixN]
                raiz = int(sqrt(len(matrixNumeros)))
                for i in range(raiz):
                    fila = matrixNumeros[i * raiz : (i + 1) * raiz]
                    cuadricula.append(fila)

                descifrado = descifrar(cuadriculaCifrada, cuadricula)
                print("Mensaje descifrado: ", descifrado)

            else:
                print("Oops!  Número no válido.")

    except ValueError:
        print("Oops!  Número no válido.")

if __name__ == "__main__":
    main()
```
