```python
def cifrado():
  mensaje = input("Escriba el mensaje: ")
  k = input("Escriba el parametro k: ")
  m = 5
  mensaje = mensaje.lower()
  mensaje = mensaje.replace(" ", "")
  codificado = ""
  resultado = ""

  for caracteres in mensaje:
    cod = ord(caracteres)+int(k)
    if cod > 122:
      cod = 97 + (cod-123)
    codificado = codificado + (chr(cod))

  for i in range(0, len(codificado), m):
        resultado += codificado[i:i+m] + " "

  codificado = resultado
  print("El mensaje cifrado es: " + codificado)

def descifrado():

  mensaje = input("Escriba el mensaje: ")
  k = input("Escriba el parametro k: ")
  m = 5
  mensaje = mensaje.lower()
  mensaje = mensaje.replace(" ", "")
  codificado = ""
  resultado = ""

  for caracteres in mensaje:
    cod = ord(caracteres)-int(k)
    if cod < 97:
      cod = 122 - (96-cod)
    codificado = codificado + (chr(cod))
  print("El mensaje cifrado es: " + codificado)

try:
  opcion = 1
  while opcion != 0:
    opcion = input("Escriba 1 para cifrar, 2 para descifrar o 0 para terminar: ")
    if int(opcion) == 0:
      break
    elif int(opcion) == 1:
      cifrado()
    elif int(opcion) == 2:
      descifrado()
    else:
      print("Oops!  Numero no valido.")


except ValueError:
  print("Oops!  Numero no valido.")
```
