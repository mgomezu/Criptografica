```python
A = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
opcion = 1
try:
  while opcion != 0:
    opcion = input("Escriba 1 para codificar, 2 para decodificar o 0 para terminar: ")
    if int(opcion) == 0:
      break
    elif int(opcion) == 1:
      mensaje = input("Escriba el mensaje: ")
      k= input("Escriba la llave: ")
      if len(mensaje) == len(k):
        mensaje = mensaje.lower()
        plaintext = []
        key = []
        ciphertextpos = []
        ciphertext = ""

        for letra in mensaje:
          if letra in A:
            plaintext.append(A.index(letra))
          else:
            plaintext.append(letra)

        for letra in k:
          key.append(A.index(letra))

        for i in range(len(mensaje)):
          if isinstance(plaintext[i], str):
            ciphertextpos.append(plaintext[i])
          else:
            cipher = (plaintext[i]+key[i])%26
            ciphertextpos.append(cipher)

        for pos in ciphertextpos:
          if isinstance(pos, str):
            ciphertext = ciphertext + pos
          else:
            ciphertext = ciphertext + A[pos]

        print("mensaje cifrado: " + ciphertext)

      else:
        print("Error, tamaño de mensaje y de llave deben ser el mismo")

    elif int(opcion) == 2:
      ciphertext = input("Escriba el mensaje cifrado: ")
      k= input("Escriba la llave: ")
      if len(ciphertext) == len(k):
        ciphertext = ciphertext.lower()
        pos = []
        plaintext = ""
        key = []
        ciphertextpos = []

        for letra in ciphertext:
          if letra in A:
            ciphertextpos.append(A.index(letra))
          else:
            ciphertextpos.append(letra)


        for letra in k:
          key.append(A.index(letra))

        for i in range(len(ciphertext)):
          if isinstance(ciphertextpos[i], str):
            pos.append(ciphertextpos[i])
          else:
            cipher = (ciphertextpos[i]-key[i])%26
            pos.append(cipher)

        for c in pos:
          if isinstance(c, str):
            plaintext = plaintext + c
          else:
            plaintext = plaintext + A[c]

        print(plaintext)

      else:
        print("Error, tamaño de mensaje y de llave deben ser el mismo")

    else:
      print("Oops!  Numero no valido.")

except ValueError:
  print("Oops!  Numero no valido.")
```
